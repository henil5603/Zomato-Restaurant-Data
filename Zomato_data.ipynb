```python
df['rate (out of 5)'].fillna(df['rate (out of 5)'].mean(),inplace=True)
```

```python
df['avg cost (two people)'].fillna(df['avg cost (two people)'].median(),inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,4:7].values
print(x)
```

```python
y = df.iloc[:,11].values
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=108)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_y = LabelEncoder()
Y=le_y.fit_transform(y)
Y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import precision_score
decision_tree_precision = precision_score(y_test,y_pred,average="weighted")
print(decision_tree_precision)
```

```python
from sklearn.metrics import recall_score
```

```python
decision_tree_recall = recall_score(y_test,y_pred,average="weighted")
print(decision_tree_recall)
```

```python
from sklearn.metrics import f1_score
```

```python
decision_tree_f1 = f1_score(y_test,y_pred,average='weighted')
print(decision_tree_f1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
